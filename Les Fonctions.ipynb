## Tri bulle

Faire une fonction tri_bulle(array) et qui tri cette liste "sur place" (en la modifiant)

Pour cela, elle compare chaque élément avec le suivant et les inverse si nécessaire.



```python
def tri_bulle(arr):
    for j in range(len(arr)-1):
        for i in range(len(arr)-1):
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
    return arr

tri_bulle([9,1,4,7,3,0])
```

## Tri fusion

Faire une fonction fusion(arr1, arr2) et une fonction tri_fusion

La fonction fusion prend deux listes triées et renvoie la fusion de ces deux listes triées

La fonction tri_fusion(arr) trie la liste si elle a moins de deux éléments (<=2). Sinon fait une fusion de deux sous tableaux triés par elle même. 

```python
def fusion(arr1, arr2):
    i = 0
    j = 0
    res = []
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    return res

fusion([1,5,7,9], [2,4,8])
```

```python
def tri_fusion(arr):
    if len(arr) < 2:
        return arr
    if len(arr) == 2:
        if arr[0] > arr[1]:
            return arr[::-1]
        return arr
    if len(arr) > 2:
        arr1 = arr[:len(arr)//2] # première moitiée
        arr2 = arr[len(arr)//2:] # deuxième moitiée
        return fusion(tri_fusion(arr1), tri_fusion(arr2))
```
